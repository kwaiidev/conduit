```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import xgboost as xgb
import pickle

# ================================
# Load data
# ================================
df = pd.read_csv('eeg_features.csv')

df.loc[600:899, "label"] = 2
df.loc[900:1199, "label"] = 3

print(f"Total samples: {len(df)}")
print(f"Features: {df.shape[1] - 1}")
print(f"Class distribution:\n{df['label'].value_counts().sort_index()}\n")

# ================================
# Chronological split PER LABEL
# Avoids: (1) data leakage from overlapping sliding windows
#         (2) test set being dominated by the last recorded label
# ================================
train_parts, test_parts = [], []

for label in sorted(df['label'].unique()):
    label_df = df[df['label'] == label].reset_index(drop=True)
    split = int(len(label_df) * 0.8)
    train_parts.append(label_df.iloc[:split])
    test_parts.append(label_df.iloc[split:])
    print(f"Label {int(label)}: {len(label_df)} total -> "
          f"{split} train / {len(label_df) - split} test")

train_df = pd.concat(train_parts).reset_index(drop=True)
test_df  = pd.concat(test_parts).reset_index(drop=True)

# ================================
# Balance TRAIN only by undersampling
# Test set is left as-is to reflect real-world conditions
# ================================
min_train = train_df['label'].value_counts().min()
balanced_parts = [
    group.sample(min_train, random_state=42)
    for _, group in train_df.groupby('label')
]
train_df = (
    pd.concat(balanced_parts)
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)

print(f"\nAfter balancing:")
print(f"  Train: {len(train_df)} samples | {train_df['label'].value_counts().sort_index().to_dict()}")
print(f"  Test:  {len(test_df)} samples  | {test_df['label'].value_counts().sort_index().to_dict()}\n")

X_train = train_df.drop('label', axis=1).values
y_train = train_df['label'].values.astype(int)
X_test  = test_df.drop('label', axis=1).values
y_test  = test_df['label'].values.astype(int)

# ================================
# Train XGBoost
# - softprob instead of softmax so we get predict_proba() in inference
# - early stopping on a validation slice to avoid overfitting
# ================================
# Carve a small validation set from the END of train (chronologically safe)
val_split  = int(len(X_train) * 0.9)
X_tr, X_val = X_train[:val_split], X_train[val_split:]
y_tr, y_val = y_train[:val_split], y_train[val_split:]

model = xgb.XGBClassifier(
    n_estimators=500,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    random_state=42,
    objective='multi:softprob',   # gives predict_proba() — required by inference.py
    num_class=len(np.unique(y_train)),
    eval_metric='mlogloss',
    early_stopping_rounds=30,     # stop if val loss doesn't improve for 30 rounds
)

print("Training...")
model.fit(
    X_tr, y_tr,
    eval_set=[(X_val, y_val)],
    verbose=50,
)
print(f"Best iteration: {model.best_iteration}\n")

# ================================
# Evaluate
# ================================
y_pred_train = model.predict(X_train)
y_pred_test  = model.predict(X_test)

train_acc = accuracy_score(y_train, y_pred_train)
test_acc  = accuracy_score(y_test,  y_pred_test)

print("=" * 55)
print("RESULTS")
print("=" * 55)
print(f"Train accuracy : {train_acc:.4f} ({train_acc*100:.2f}%)")
print(f"Test  accuracy : {test_acc:.4f}  ({test_acc*100:.2f}%)")

gap = train_acc - test_acc
if gap > 0.10:
    print(f"  ⚠ Large train/test gap ({gap:.2f}) — model may be overfitting.")
elif test_acc < 0.60:
    print(f"  ⚠ Low test accuracy — consider collecting more data per label.")
else:
    print(f"  ✓ Train/test gap looks healthy ({gap:.2f}).")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_test)
n_classes = cm.shape[0]
print(f"\nConfusion Matrix (rows=actual, cols=predicted):")
header = "          " + "".join(f"  Pred {i}" for i in range(n_classes))
print(header)
for i in range(n_classes):
    row = f"  Act {i}  " + "".join(f"  {cm[i][j]:>6}" for j in range(n_classes))
    print(row)

print(f"\nClassification Report:")
print(classification_report(
    y_test, y_pred_test,
    target_names=[f'Class {i}' for i in range(n_classes)]
))

# Per-class accuracy
print("Per-Class Accuracy:")
for cls in range(n_classes):
    mask = y_test == cls
    if mask.sum() > 0:
        cls_acc = accuracy_score(y_test[mask], y_pred_test[mask])
        print(f"  Class {cls}: {cls_acc:.4f} ({cls_acc*100:.2f}%)  [{mask.sum()} samples]")

# Feature importance
feature_cols = train_df.drop('label', axis=1).columns
feat_imp = (
    pd.DataFrame({'feature': feature_cols, 'importance': model.feature_importances_})
    .sort_values('importance', ascending=False)
    .reset_index(drop=True)
)
print(f"\nTop 15 Features:")
print(feat_imp.head(15).to_string(index=False))

# ================================
# Save
# ================================
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
print(f"\n✓ Model saved to model.pkl")
print(f"  Trained on {len(X_train)} samples, tested on {len(X_test)} samples")
print(f"  Feature vector length: {X_train.shape[1]}")
```
